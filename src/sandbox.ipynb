```python
from prophandler import PropTracker
tracker = PropTracker()
```

```python
tracker.data().sort_values(['just_moved', 'e_movements'], ascending=False).head(10)
```

```python
import datetime
import pandas as pd
import settings.custom
def visualize(name: str, value: str = 'e_props') -> pd.DataFrame:
    df = tracker.data().loc[name] #, ['props', 'e_props']]

    n_props = len(list(df.props))
    
    df_viz = pd.DataFrame(data = {
        'name': [name]*n_props,
        'props': list(df.props),
        'e_props': list(df.e_props),
        'scrape_times': list(df.scrape_times)
    })

    step = (max(df_viz[value]) - min(df_viz[value])) / 2

    
    
    return df_viz.groupby('name')[value].plot.line(
        title=f'{value} movement for {name} since {df.init_time}',
        figsize=(15,7),
        xticks=[i for i, time in enumerate(df_viz.scrape_times) if not i % 5] if n_props > 30 else [i for i, time in enumerate(df_viz.scrape_times) if not i % 1],
        yticks=[round(min(df_viz[value]) + n*step, 2) for n in range(-1, n_props+10) if (min(df_viz[value]) + n*step) <= max(df_viz[value])+step]
    )
```

```python
def graph(name: str):
    visualize(name)
    # visualize(name, value='props')
    return
```
